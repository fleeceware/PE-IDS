```python
import pandas as pd
```

```python
# Creating a DataFrame
data = {
    'Name': ['Amit', 'Riya', 'Sohan', 'Tina'],
    'Age': [21, 23, 20, 22],
    'Marks': [85, 90, 75, 95]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# Selecting specific column
print("\nNames Column:\n", df['Name'])
```

```python
# Filtering data
print("\nStudents with Marks > 80:\n", df[df['Marks'] > 80])
```

```python
# Adding a new column
df['Result'] = ['Pass' if x >= 40 else 'Fail' for x in df['Marks']]
print("\nDataFrame after adding new column:\n", df)
```

```python
# Sorting data
print("\nData sorted by Marks:\n", df.sort_values(by='Marks', ascending=False))
```

```python
# Aggregation
print("\nAverage Marks:", df['Marks'].mean())
```

```python
# Deleting a column
df = df.drop('Age', axis=1)
print("\nDataFrame after deleting 'Age' column:\n", df)
```
