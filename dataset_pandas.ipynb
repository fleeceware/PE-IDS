```python
import pandas as pd
```

```python
# Loading dataset (example dataset)
data = {
    'Name': ['Amit', 'Riya', 'Sohan', 'Tina', 'Rahul'],
    'Age': [21, 23, 20, 22, 21],
    'Department': ['IT', 'CS', 'IT', 'ENTC', 'CS'],
    'Marks': [85, 90, 75, 95, 88]
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)
```

```python
# Display first few records
print("\nFirst 3 records of dataset:\n", df.head(3))
```

```python
# Information about dataset
print("\nDataset Information:\n")
print(df.info())
```

```python
# Describe numerical columns
print("\nStatistical Summary:\n", df.describe())
```

```python
# Filtering students with Marks > 80
print("\nStudents with Marks greater than 80:\n", df[df['Marks'] > 80])
```

```python
# Sorting data by Marks
sorted_df = df.sort_values(by='Marks', ascending=False)
print("\nDataset sorted by Marks (Descending Order):\n", sorted_df)
```

```python
# Grouping data by Department
grouped_data = df.groupby('Department')['Marks'].mean()
print("\nAverage Marks by Department:\n", grouped_data)
```

```python
 #Adding a new column (Result)
df['Result'] = ['Pass' if x >= 40 else 'Fail' for x in df['Marks']]
print("\nDataset after adding 'Result' column:\n", df)
```

```python
# Removing a column
df = df.drop('Age', axis=1)
print("\nDataset after deleting 'Age' column:\n", df)
```

```python
# Handling missing values example
df.loc[2, 'Marks'] = None
print("\nDataset with Missing Value Added:\n", df)
```

```python
 #Filling missing values
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
print("\nDataset after Filling Missing Value with Mean:\n", df)
```
